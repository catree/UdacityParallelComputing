# Uniform Blur vs. Gaussian Blur

The purpose of this code is to perform a uniform blur and a Gaussian Blur using the same kernel size. We see that the results are different, especially around strong image edges.

## load image

```python
import PIL
import PIL.Image

image                 = PIL.Image.open("cinque_terre.jpg")
image_array_rgb       = numpy.array(image)

r,g,b                 = numpy.split(image_array_rgb, 3, axis=2)
a                     = numpy.ones_like(r) * 255

image_array_rgba      = numpy.concatenate((r,g,b,a), axis=2).copy()



figsize(6,4)

matplotlib.pyplot.imshow(image_array_rgba);
matplotlib.pyplot.title("image_array_rgba");
```

## isolate small greyscale patch from the original image

```python
r_small = r[1150:1450,2300:2600,0].copy().astype(numpy.uint8)



figsize(4,4)

matplotlib.pyplot.imshow(r_small, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small");
```

## common uniform blur code

```python
uniform_blur_kernel_width      = numpy.int32(9)
uniform_blur_kernel_half_width = numpy.int32(4)
uniform_blur_kernel_value      = numpy.float32(1) / numpy.float32(uniform_blur_kernel_width**2)
uniform_blur_kernel            = numpy.ones((uniform_blur_kernel_width, uniform_blur_kernel_width), dtype=numpy.float32) * uniform_blur_kernel_value
```

## CPU uniform blur

```python
import scipy
import scipy.ndimage
import scipy.ndimage.filters

convolution_filter          = numpy.ones((uniform_blur_kernel_width, uniform_blur_kernel_width), dtype=numpy.float32) * uniform_blur_kernel_value
r_small_blurred_cpu_uniform = scipy.ndimage.filters.convolve(r_small, convolution_filter, mode="nearest")



figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_cpu_uniform, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu_uniform");
```

## common gaussian blur code

We begin by defining the non-normalized Gaussian Function $G(x,y)$ as follows:

### $$G(x,y) = e^{ - \frac{ x^2 + y^2 }{ 2 { \sigma }^2 } }$$

We then define the normalized Gaussian Function as $N(x,y)=cG(x,y)$, where $c$ is a normalization constant:

### $$c = \frac{ 1 }{ \int{ \int{ G(x,y) } } \mathrm{d}x \mathrm{d}y }$$

To perform a Gaussian Blur, we use $N(x,y)$ as a convolution kernel.

```python
import scipy

gaussian_blur_kernel_width      = numpy.int32(9)
gaussian_blur_kernel_half_width = numpy.int32(4)
gaussian_blur_sigma             = numpy.float32(2)

y, x = \
    scipy.mgrid[-gaussian_blur_kernel_half_width:gaussian_blur_kernel_half_width+1,-gaussian_blur_kernel_half_width:gaussian_blur_kernel_half_width+1]

gaussian_blur_kernel_not_normalized = numpy.exp( ( - ( x**2 + y**2 ) ) / ( 2 * gaussian_blur_sigma**2 ) )
normalization_constant              = numpy.float32(1) / numpy.sum(gaussian_blur_kernel_not_normalized)
gaussian_blur_kernel                = (normalization_constant * gaussian_blur_kernel_not_normalized).astype(numpy.float32)



figsize(4,4)

matplotlib.pyplot.imshow(gaussian_blur_kernel, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("gaussian_blur_kernel");
matplotlib.pyplot.colorbar();
```

## CPU gaussian blur

```python
import scipy.ndimage
import scipy.ndimage.filters

convolution_filter           = gaussian_blur_kernel
r_small_blurred_cpu_gaussian = scipy.ndimage.filters.convolve(r_small, convolution_filter, mode="nearest")



figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_cpu_gaussian, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu_gaussian");
```

## CPU uniform blur vs. CPU gaussian blur

```python
diff = numpy.abs(r_small_blurred_cpu_uniform.astype(numpy.float32) - r_small_blurred_cpu_gaussian.astype(numpy.float32))



figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_cpu_uniform, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu_uniform")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_cpu_gaussian, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu_gaussian")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```
