# GPU Split Test

The purpose of this code is to implement a canonical split algorithm on the GPU.

## generate data

```python
import split

n                    = 10000
input_data           = (numpy.random.rand(n) * n).astype(numpy.uint32)
flag_data            = numpy.zeros_like(input_data)

b    = 0
mask = 2**b

for i in range(n):
    input_value  = input_data[i]
    flag_data[i] = not (input_value & mask)

print input_data
print flag_data
print

figsize(9,4)

matplotlib.pyplot.subplot(121)
matplotlib.pyplot.plot(input_data);
matplotlib.pyplot.title("input_data");

matplotlib.pyplot.subplot(122)
matplotlib.pyplot.plot(flag_data);
matplotlib.pyplot.title("flag_data");
```

## CPU split

```python
flag_set_scatter_offset_cpu = zeros_like(input_data)
output_data_cpu             = zeros_like(input_data)

for i in range(1,n):
    flag_set_scatter_offset_cpu[i] = flag_set_scatter_offset_cpu[i-1] + flag_data[i-1]

total_flags_set = flag_set_scatter_offset_cpu[n-1] + flag_data[n-1]

for i in range(n):
    input_value                   = input_data[i]
    flag_value                    = flag_data[i]
    flag_set_scatter_offset_value = flag_set_scatter_offset_cpu[i]
    scatter_offset_value          = -1

    if flag_value:
        scatter_offset_value = flag_set_scatter_offset_value
    else:
        scatter_offset_value = i - flag_set_scatter_offset_value + total_flags_set

    output_data_cpu[scatter_offset_value] = input_value



print input_data
print flag_data
print flag_set_scatter_offset_cpu
print output_data_cpu
print

figsize(19,4)

matplotlib.pyplot.subplot(141)
matplotlib.pyplot.plot(input_data);
matplotlib.pyplot.title("input_data");

matplotlib.pyplot.subplot(142)
matplotlib.pyplot.plot(flag_data);
matplotlib.pyplot.title("flag_data");

matplotlib.pyplot.subplot(143)
matplotlib.pyplot.plot(flag_set_scatter_offset_cpu);
matplotlib.pyplot.title("flag_set_scatter_offset_cpu");

matplotlib.pyplot.subplot(144)
matplotlib.pyplot.plot(output_data_cpu);
matplotlib.pyplot.title("output_data_cpu");
```

## GPU split

```python
import pycuda.autoinit
import pycuda.driver
import pycuda.compiler
import prefixsum

source_module = pycuda.compiler.SourceModule \
(
"""
__global__ void split_scatter(
    unsigned int* d_input_data,
    unsigned int* d_flag_data,
    unsigned int* d_flag_set_scatter_offset,
    unsigned int* d_output_data,
    int total_flags_set,
    int n )
{
    int global_index_1d = ( blockIdx.x * blockDim.x ) + threadIdx.x;

    if ( global_index_1d < n )
    {
        unsigned int input_value                   = d_input_data[ global_index_1d ];
        unsigned int flag_value                    = d_flag_data[ global_index_1d ];
        unsigned int flag_set_scatter_offset_value = d_flag_set_scatter_offset[ global_index_1d ];
        
        unsigned int scatter_offset_value;

        if ( flag_value > 0 )
        {
            scatter_offset_value = flag_set_scatter_offset_value;
        }
        else
        {
            scatter_offset_value = global_index_1d - flag_set_scatter_offset_value + total_flags_set;
        }
    
        d_output_data[ scatter_offset_value ] = input_value;
    }
}
"""
)

split_scatter_funcion = source_module.get_function("split_scatter")

size_of_element_bytes = 4
max_num_elements      = 15000
num_bytes             = max_num_elements * size_of_element_bytes

input_data_device              = pycuda.driver.mem_alloc(num_bytes)
flag_data_device               = pycuda.driver.mem_alloc(num_bytes)
flag_set_scatter_offset_device = pycuda.driver.mem_alloc(num_bytes)
output_data_device             = pycuda.driver.mem_alloc(num_bytes)

prefix_sum_manager = prefixsum.PrefixSumManager(max_num_elements)
```

```python
pycuda.driver.memcpy_htod(input_data_device, input_data)
pycuda.driver.memcpy_htod(flag_data_device,  flag_data)

prefix_sum_manager.prefix_sum_device(flag_data_device, flag_set_scatter_offset_device, n)

tmp = numpy.zeros(1, dtype=numpy.uint32)

pycuda.driver.memcpy_dtoh(tmp,int(flag_set_scatter_offset_device) + ((n-1) * size_of_element_bytes))
flag_set_scatter_offset_end = tmp[0]
pycuda.driver.memcpy_dtoh(tmp,int(flag_data_device) + ((n-1) * size_of_element_bytes))
flag_data_end = tmp[0]

total_flags_set = flag_set_scatter_offset_end + flag_data_end

split_scatter_funcion_block = (512,1,1)
num_blocks                  = int(ceil(float(n) / float(split_scatter_funcion_block[0])))
split_scatter_function_grid = (num_blocks, 1)

split_scatter_funcion(
    input_data_device,
    flag_data_device,
    flag_set_scatter_offset_device,
    output_data_device,
    numpy.int32(total_flags_set),
    numpy.int32(n),
    block=split_scatter_funcion_block,
    grid=split_scatter_function_grid)

output_data_gpu = numpy.zeros_like(input_data)
pycuda.driver.memcpy_dtoh(output_data_gpu, output_data_device)



print output_data_gpu
print output_data_cpu
print
print "Difference between GPU and CPU (should be 0.0%%): %f" % numpy.linalg.norm(output_data_gpu - output_data_cpu)

figsize(19,4)

matplotlib.pyplot.subplot(141)
matplotlib.pyplot.plot(input_data);
matplotlib.pyplot.title("input_data");

matplotlib.pyplot.subplot(142)
matplotlib.pyplot.plot(flag_data);
matplotlib.pyplot.title("flag_data");

matplotlib.pyplot.subplot(143)
matplotlib.pyplot.plot(output_data_gpu);
matplotlib.pyplot.title("output_data_gpu");

matplotlib.pyplot.subplot(144)
matplotlib.pyplot.plot(output_data_cpu);
matplotlib.pyplot.title("output_data_cpu");
```

## invoke GPU split class

```python
import split

n           = 10000
input_data  = (numpy.random.rand(n) * n).astype(numpy.uint32)
flag_data   = zeros_like(input_data)

b    = 0
mask = 2**b
    
for i in range(n):
    input_value  = input_data[i]
    flag_data[i] = not (input_value & mask)


    
flag_set_scatter_offset_cpu = zeros_like(input_data)
output_data_cpu             = zeros_like(input_data)

for i in range(1,n):
    flag_set_scatter_offset_cpu[i] = flag_set_scatter_offset_cpu[i-1] + flag_data[i-1]

total_flags_set = flag_set_scatter_offset_cpu[n-1] + flag_data[n-1]

for i in range(n):
    input_value                   = input_data[i]
    flag_value                    = flag_data[i]
    flag_set_scatter_offset_value = flag_set_scatter_offset_cpu[i]
    scatter_offset_value          = -1

    if flag_value:
        scatter_offset_value = flag_set_scatter_offset_value
    else:
        scatter_offset_value = i - flag_set_scatter_offset_value + total_flags_set

    output_data_cpu[scatter_offset_value] = input_value
    
    

output_data_gpu = numpy.zeros_like(input_data)
split_manager   = split.SplitManager(15000)

split_manager.split_host(input_data, flag_data, output_data_gpu)

print output_data_gpu
print output_data_cpu
print
print "Difference between GPU and CPU (should be 0.0%%): %f" % numpy.linalg.norm(output_data_gpu - output_data_cpu)

figsize(19,4)

matplotlib.pyplot.subplot(141)
matplotlib.pyplot.plot(input_data);
matplotlib.pyplot.title("input_data");

matplotlib.pyplot.subplot(142)
matplotlib.pyplot.plot(flag_data);
matplotlib.pyplot.title("flag_data");

matplotlib.pyplot.subplot(143)
matplotlib.pyplot.plot(output_data_gpu);
matplotlib.pyplot.title("output_data_gpu");

matplotlib.pyplot.subplot(144)
matplotlib.pyplot.plot(output_data_cpu);
matplotlib.pyplot.title("output_data_cpu");
```
