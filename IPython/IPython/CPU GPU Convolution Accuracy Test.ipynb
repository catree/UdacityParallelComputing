# CPU GPU Convolution Test

This code compares the results from various convolution methods on the CPU and GPU. We find that the GPU results are not identical to the CPU, regardless of the CPU method used. See [1] for more details.

[1] http://developer.download.nvidia.com/CUDA/training/floatingpointwebinarjune2011.pdf 

## load image

```python
import PIL
import PIL.Image

image                 = PIL.Image.open("cinque_terre.jpg")
image_array_rgb       = numpy.array(image)

r,g,b                 = numpy.split(image_array_rgb, 3, axis=2)
a                     = numpy.ones_like(r) * 255

image_array_rgba      = numpy.concatenate((r,g,b,a), axis=2).copy()


figsize(6,4)

matplotlib.pyplot.imshow(image_array_rgba);
matplotlib.pyplot.title("image_array_rgba");
```

## isolate small greyscale patch from the original image

```python
r_small = r[1150:1450,2300:2600,0].copy().astype(numpy.uint8)



figsize(4,4)

matplotlib.pyplot.imshow(r_small, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small");
```

## common uniform blur code

```python
uniform_blur_kernel_width      = numpy.int32(9)
uniform_blur_kernel_half_width = numpy.int32(4)
uniform_blur_kernel_value      = numpy.float32(1) / numpy.float32(uniform_blur_kernel_width**2)
```

## CPU uniform blur hand-coded loop

```python
r_small_blurred_loop = numpy.zeros_like(r_small)

num_pixels_y         = r_small.shape[0]
num_pixels_x         = r_small.shape[1]

for y in range(num_pixels_y):
    for x in range(num_pixels_x):
    
        result = numpy.float32(0)
    
        for offset_y in range(-uniform_blur_kernel_half_width, uniform_blur_kernel_half_width + 1):
            for offset_x in range(-uniform_blur_kernel_half_width, uniform_blur_kernel_half_width + 1):
                
                index_y         = y + offset_y
                index_x         = x + offset_x
                
                index_y_clamped = min(num_pixels_y - 1, max(0, index_y)) 
                index_x_clamped = min(num_pixels_x - 1, max(0, index_x)) 
    
                result          = result + (uniform_blur_kernel_value * r_small[index_y_clamped,index_x_clamped])
    
        r_small_blurred_loop[y,x] = numpy.uint8(result)
        
        

figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_loop, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_loop");
```

## CPU uniform blur hand-coded loop vs. scipy.ndimage.filters.uniform_filter

```python
import scipy
import scipy.ndimage
import scipy.ndimage.filters

r_small_blurred_uniform_filter = scipy.ndimage.filters.uniform_filter(r_small, 9, mode="nearest")
diff                           = numpy.abs(r_small_blurred_loop.astype(float32) - r_small_blurred_uniform_filter.astype(float32))
max_diff                       = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between hand-coded CPU loop and scipy.ndimage.filters.uniform_filter results " + \
    "as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%"    % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_loop, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_loop")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_uniform_filter, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_uniform_filter")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## CPU uniform blur hand-coded loop vs. scipy.ndimage.filters.convolve

```python
convolution_filter       = numpy.ones((uniform_blur_kernel_width, uniform_blur_kernel_width), dtype=numpy.float32) * uniform_blur_kernel_value
r_small_blurred_convolve = scipy.ndimage.filters.convolve(r_small, convolution_filter, mode="nearest")
diff                     = numpy.abs(r_small_blurred_loop.astype(float32) - r_small_blurred_convolve.astype(float32))
max_diff                 = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between hand-coded CPU loop and scipy.ndimage.filters.convolve results "    + \
    "as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_loop, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_loop")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_convolve, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_convolve")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## CPU uniform blur hand-coded loop vs. scipy.ndimage.filters.generic_filter

```python
def convolution_function(buffer):
    result = numpy.float32(0)

    for i in range(buffer.shape[0]):
        result = result + (uniform_blur_kernel_value * buffer[i])

    return result

r_small_blurred_generic_filter = scipy.ndimage.filters.generic_filter(r_small, convolution_function, size=9, mode="nearest")
diff                           = numpy.abs(r_small_blurred_loop.astype(float32) - r_small_blurred_generic_filter.astype(float32))
max_diff                       = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between hand-coded CPU loop and scipy.ndimage.filters.generic_filter "      + \
    "as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_loop, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_loop")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_generic_filter, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_generic_filter")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## scipy.ndimage.filters.convolve vs. scipy.ndimage.filters.generic_filter

```python
diff     = numpy.abs(r_small_blurred_convolve.astype(float32) - r_small_blurred_generic_filter.astype(float32))
max_diff = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between scipy.ndimage.filters.convolve and scipy.ndimage.filters.generic_filter results " + \
    "as a percentage of the maximum possible difference (should be 0%%): %0.2f%%"                         % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_convolve, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_convolve")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_generic_filter, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_generic_filter")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## CPU uniform blur hand-coded loop vs. GPU naive uniform blur (unsigned char)result

```python
import pycuda.autoinit
import pycuda.driver
import pycuda.compiler

r_small_blurred_gpu_naive_unsigned_char = numpy.zeros_like(r_small)

source_module = pycuda.compiler.SourceModule \
(
"""
__global__ void naive_uniform_blur(
    unsigned char* d_blurred,
    unsigned char* d_original,
    int            num_pixels_y,
    int            num_pixels_x,
    int            blur_kernel_half_width,
    float          blur_kernel_value )
{
    int  ny             = num_pixels_y;
    int  nx             = num_pixels_x;
    int2 image_index_2d = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int  image_index_1d = ( nx * image_index_2d.y ) + image_index_2d.x;

    if ( image_index_2d.x < nx && image_index_2d.y < ny )
    {
        float result = 0;

        for ( int y = -blur_kernel_half_width; y <= blur_kernel_half_width; y++ )
        {
            for ( int x = -blur_kernel_half_width; x <= blur_kernel_half_width; x++ )
            {
                int2 image_offset_index_2d         = make_int2( image_index_2d.x + x, image_index_2d.y + y );
                int2 image_offset_index_2d_clamped = make_int2( min( nx - 1, max( 0, image_offset_index_2d.x ) ), min( ny - 1, max( 0, image_offset_index_2d.y ) ) );
                int  image_offset_index_1d_clamped = ( nx * image_offset_index_2d_clamped.y ) + image_offset_index_2d_clamped.x;
    
                result += blur_kernel_value * d_original[ image_offset_index_1d_clamped ];
            }
        }

        d_blurred[ image_index_1d ] = (unsigned char)result;
    }
}
"""
)

naive_uniform_blur_function       = source_module.get_function("naive_uniform_blur")

num_pixels_y                      = numpy.int32(r_small.shape[0])
num_pixels_x                      = numpy.int32(r_small.shape[1])
naive_uniform_blur_function_block = (32,16,1)

num_blocks_y                      = int(ceil(float(num_pixels_y) / float(naive_uniform_blur_function_block[1])))
num_blocks_x                      = int(ceil(float(num_pixels_x) / float(naive_uniform_blur_function_block[0])))
naive_uniform_blur_function_grid  = (num_blocks_x, num_blocks_y)

r_small_device                    = pycuda.driver.mem_alloc(r_small.nbytes)
r_small_blurred_device            = pycuda.driver.mem_alloc(r_small_blurred_gpu_naive_unsigned_char.nbytes)

pycuda.driver.memcpy_htod(r_small_device,         r_small)
pycuda.driver.memcpy_htod(r_small_blurred_device, r_small_blurred_gpu_naive_unsigned_char)

naive_uniform_blur_function(
    r_small_blurred_device,
    r_small_device,
    num_pixels_y,
    num_pixels_x,
    uniform_blur_kernel_half_width,
    uniform_blur_kernel_value,
    block=naive_uniform_blur_function_block,
    grid=naive_uniform_blur_function_grid)

pycuda.driver.memcpy_dtoh(r_small_blurred_gpu_naive_unsigned_char, r_small_blurred_device)

diff     = numpy.abs(r_small_blurred_loop.astype(float32) - r_small_blurred_gpu_naive_unsigned_char.astype(float32))
max_diff = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between hand-coded CPU loop and GPU '(unsigned char)result' results "       + \
    "as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_loop, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_loop")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_gpu_naive_unsigned_char, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_naive_unsigned_char")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## CPU uniform blur hand-coded loop vs. GPU naive uniform blur __float2uint_rd(result)

```python
import pycuda.autoinit
import pycuda.driver
import pycuda.compiler

r_small_blurred_gpu_naive_float2uint = numpy.zeros_like(r_small)

source_module = pycuda.compiler.SourceModule \
(
"""
__global__ void naive_uniform_blur(
    unsigned char* d_blurred,
    unsigned char* d_original,
    int            num_pixels_y,
    int            num_pixels_x,
    int            blur_kernel_half_width,
    float          blur_kernel_value )
{
    int  ny             = num_pixels_y;
    int  nx             = num_pixels_x;
    int2 image_index_2d = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int  image_index_1d = ( nx * image_index_2d.y ) + image_index_2d.x;

    if ( image_index_2d.x < nx && image_index_2d.y < ny )
    {
        float result = 0;

        for ( int y = -blur_kernel_half_width; y <= blur_kernel_half_width; y++ )
        {
            for ( int x = -blur_kernel_half_width; x <= blur_kernel_half_width; x++ )
            {
                int2 image_offset_index_2d         = make_int2( image_index_2d.x + x, image_index_2d.y + y );
                int2 image_offset_index_2d_clamped = make_int2( min( nx - 1, max( 0, image_offset_index_2d.x ) ), min( ny - 1, max( 0, image_offset_index_2d.y ) ) );
                int  image_offset_index_1d_clamped = ( num_pixels_x * image_offset_index_2d_clamped.y ) + image_offset_index_2d_clamped.x;
    
                result += blur_kernel_value * d_original[ image_offset_index_1d_clamped ];
            }
        }

        d_blurred[ image_index_1d ] = __float2uint_rd( result );
    }
}
"""
)

naive_uniform_blur_function       = source_module.get_function("naive_uniform_blur")

num_pixels_y                      = numpy.int32(r_small.shape[0])
num_pixels_x                      = numpy.int32(r_small.shape[1])
naive_uniform_blur_function_block = (32,16,1)

num_blocks_y                      = int(ceil(float(num_pixels_y) / float(naive_uniform_blur_function_block[1])))
num_blocks_x                      = int(ceil(float(num_pixels_x) / float(naive_uniform_blur_function_block[0])))
naive_uniform_blur_function_grid  = (num_blocks_x, num_blocks_y)

r_small_device                    = pycuda.driver.mem_alloc(r_small.nbytes)
r_small_blurred_device            = pycuda.driver.mem_alloc(r_small_blurred_gpu_naive_float2uint.nbytes)

pycuda.driver.memcpy_htod(r_small_device,         r_small)
pycuda.driver.memcpy_htod(r_small_blurred_device, r_small_blurred_gpu_naive_float2uint)

naive_uniform_blur_function(
    r_small_blurred_device,
    r_small_device,
    num_pixels_y,
    num_pixels_x,
    uniform_blur_kernel_half_width,
    uniform_blur_kernel_value,
    block=naive_uniform_blur_function_block,
    grid=naive_uniform_blur_function_grid)

pycuda.driver.memcpy_dtoh(r_small_blurred_gpu_naive_float2uint, r_small_blurred_device)

diff     = numpy.abs(r_small_blurred_loop.astype(float32) - r_small_blurred_gpu_naive_float2uint.astype(float32))
max_diff = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between hand-coded CPU loop and GPU '__float2uint_rd( result )' results " +   \
    "as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))
    
print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_loop, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_loop")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_gpu_naive_float2uint, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_naive_float2uint")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```
