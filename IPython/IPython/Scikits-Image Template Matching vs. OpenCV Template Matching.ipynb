# Scikits-Image Template Matching vs. OpenCV Template Matching

The purpose of this code is to match a template to a photograph using Scikits-Image and OpenCV. We find that the results from each library are quite different.

## load images

```python
import PIL
import PIL.Image

image                            = PIL.Image.open("red_eye_effect_5.jpg")
image_template                   = PIL.Image.open("red_eye_effect_template_5.jpg")

image_array_rgb                  = numpy.array(image)
image_template_array_rgb         = numpy.array(image_template)

r,g,b                            = numpy.split(image_array_rgb, 3, axis=2)
r_template,g_template,b_template = numpy.split(image_template_array_rgb, 3, axis=2)

a                                = numpy.ones_like(r) * 255
a_template                       = numpy.ones_like(r_template) * 255

rgba                             = numpy.concatenate((r,g,b,a), axis=2).copy()
rgba_template                    = numpy.concatenate((r_template,g_template,b_template,a_template), axis=2).copy()

r = r.squeeze().copy()
g = g.squeeze().copy()
b = b.squeeze().copy()

r_template = r_template.squeeze().copy()
g_template = g_template.squeeze().copy()
b_template = b_template.squeeze().copy()



figsize(9,4)

matplotlib.pyplot.subplot(121)
matplotlib.pyplot.imshow(rgba);
matplotlib.pyplot.title("rgba");

matplotlib.pyplot.subplot(122)
matplotlib.pyplot.imshow(rgba_template);
matplotlib.pyplot.title("rgba_template");
```

## skimage.feature.match_template vs. cv2.matchTemplate

```python
import skimage
import skimage.feature
import cv2

image                  = r.astype(numpy.float32)          / 255.0
template               = r_template.astype(numpy.float32) / 255.0

response_skimage       = skimage.feature.match_template(image, template)
response_opencv        = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
diff                   = numpy.abs(response_skimage - response_opencv)

template_half_height   = (template.shape[0] - 1) / 2
template_half_width    = (template.shape[1] - 1) / 2

response_skimage_small = response_skimage[ 350-template_half_height:450-template_half_height, 200-template_half_width:300-template_half_width ]
response_opencv_small  = response_opencv[ 350-template_half_height:450-template_half_height, 200-template_half_width:300-template_half_width ]
diff_small             = numpy.abs(response_skimage_small - response_opencv_small)



figsize(24,14)

matplotlib.pyplot.subplot(231);
matplotlib.pyplot.imshow(response_skimage, cmap="gray");
matplotlib.pyplot.title("response_skimage");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(232);
matplotlib.pyplot.imshow(response_opencv, cmap="gray");
matplotlib.pyplot.title("response_opencv");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(233);
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(234);
matplotlib.pyplot.imshow(response_skimage_small, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("response_skimage_small");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(235);
matplotlib.pyplot.imshow(response_opencv_small, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("response_opencv_small");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(236);
matplotlib.pyplot.imshow(diff_small, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("diff_small");
matplotlib.pyplot.colorbar();
```
