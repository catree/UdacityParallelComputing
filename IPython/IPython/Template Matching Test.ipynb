# Template Matching Test

The purpose of this code is to match a template to a photograph.

## load images

```python
import PIL
import PIL.Image

image                            = PIL.Image.open("red_eye_effect_5.jpg")
image_template                   = PIL.Image.open("red_eye_effect_template_5.jpg")

image_array_rgb                  = numpy.array(image)
image_template_array_rgb         = numpy.array(image_template)

r,g,b                            = numpy.split(image_array_rgb, 3, axis=2)
r_template,g_template,b_template = numpy.split(image_template_array_rgb, 3, axis=2)

a                                = numpy.ones_like(r) * 255
a_template                       = numpy.ones_like(r_template) * 255

rgba                             = numpy.concatenate((r,g,b,a), axis=2).copy()
rgba_template                    = numpy.concatenate((r_template,g_template,b_template,a_template), axis=2).copy()

r = r.squeeze().copy()
g = g.squeeze().copy()
b = b.squeeze().copy()

r_template = r_template.squeeze().copy()
g_template = g_template.squeeze().copy()
b_template = b_template.squeeze().copy()



figsize(9,4)

matplotlib.pyplot.subplot(121)
matplotlib.pyplot.imshow(rgba);
matplotlib.pyplot.title("rgba");

matplotlib.pyplot.subplot(122)
matplotlib.pyplot.imshow(rgba_template);
matplotlib.pyplot.title("rgba_template");
```

## perform normalized cross-correlation to find good template matches

```python
import cv2

r_response_opencv = cv2.matchTemplate(r.astype(numpy.float32), r_template.astype(numpy.float32), cv2.TM_CCOEFF_NORMED)
g_response_opencv = cv2.matchTemplate(g.astype(numpy.float32), g_template.astype(numpy.float32), cv2.TM_CCOEFF_NORMED)
b_response_opencv = cv2.matchTemplate(b.astype(numpy.float32), b_template.astype(numpy.float32), cv2.TM_CCOEFF_NORMED)

response_opencv   = r_response_opencv * g_response_opencv * b_response_opencv



figsize(20,4)

matplotlib.pyplot.subplot(141);
matplotlib.pyplot.imshow(r_response_opencv, cmap="gray");
matplotlib.pyplot.title("r_response_opencv");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(142);
matplotlib.pyplot.imshow(g_response_opencv, cmap="gray");
matplotlib.pyplot.title("g_response_opencv");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(143);
matplotlib.pyplot.imshow(b_response_opencv, cmap="gray");
matplotlib.pyplot.title("b_response_opencv");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(144);
matplotlib.pyplot.imshow(response_opencv, cmap="gray");
matplotlib.pyplot.title("response_opencv");
matplotlib.pyplot.colorbar();
```

## threshold response image to find best template matches

```python
figsize(6,4)

matplotlib.pyplot.imshow(response_opencv > 0.15, cmap="gray");
matplotlib.pyplot.title("response_opencv > 0.15");
matplotlib.pyplot.colorbar();
```

## define CPU normalized cross-correlation

We use the following expression for normalized cross-correlation:

### $$ R_{x,y} = \frac{ \sum_{u,v}{( I_{u,v}( T_{u,v} - T_{\text{average}} ) )} }    { \sqrt{ \sum_{u,v}{ ( ( I_{u,v} - I_{\text{average}} )^{2} ) \sum_{u,v}{ ( ( T_{u,v} - T_{\text{average}} )^2 ) } } } } $$

Where $I_{u,v}$ is the current image intensity value under the template, $I_{\text{average}}$ is the average image intensity under the template, $T_{u,v}$ is the current template intensity value, $T_{\text{average}}$ is the average template intensity value. See [1] and [2] for more details.

[1] http://www.idiom.com/~zilla/Papers/nvisionInterface/nip.html

[2] https://github.com/scikits-image/scikits-image/blob/master/skimage/feature/_template.pyx

```python
def normalized_cross_correlation_cpu_loop(image, template):

    result               = numpy.zeros_like(image)

    num_pixels_y         = image.shape[0]
    num_pixels_x         = image.shape[1]

    template_height      = template.shape[0]
    template_width       = template.shape[1]
    
    template_half_height = (template_height - 1) / 2
    template_half_width  = (template_width  - 1) / 2
    
    template_size        = numpy.float32(template_height * template_width)
    template_mean        = numpy.mean(template)

    for y in range(num_pixels_y):
        for x in range(num_pixels_x):

            #
            # compute image mean
            #
            image_sum = numpy.float32(0)
            
            for offset_y in range(-template_half_height, template_half_height + 1):
                for offset_x in range(-template_half_width, template_half_width + 1):
                    
                    index_y         = y + offset_y
                    index_x         = x + offset_x
                    
                    index_y_clamped = min(num_pixels_y - 1, max(0, index_y)) 
                    index_x_clamped = min(num_pixels_x - 1, max(0, index_x)) 
    
                    image_value     = image[index_y_clamped,index_x_clamped]
                    image_sum       = image_sum + image_value
    
            image_mean = image_sum / template_size

            #
            # compute sums
            #
            sum_of_image_template_diff_products = numpy.float32(0)
            sum_of_squared_image_diffs          = numpy.float32(0)
            sum_of_squared_template_diffs       = numpy.float32(0)
        
            for offset_y in range(-template_half_height, template_half_height + 1):
                for offset_x in range(-template_half_width, template_half_width + 1):
                
                    index_y                     = y + offset_y
                    index_x                     = x + offset_x
                    
                    index_y_clamped             = min(num_pixels_y - 1, max(0, index_y)) 
                    index_x_clamped             = min(num_pixels_x - 1, max(0, index_x)) 
        
                    template_index_y            = offset_y + template_half_height
                    template_index_x            = offset_x + template_half_width
                
                    image_value                 = image[index_y_clamped,index_x_clamped]
                    template_value              = template[template_index_y,template_index_x]
    
                    image_diff                  = image_value    - image_mean
                    template_diff               = template_value - template_mean
                    
                    image_template_diff_product = image_value * template_diff
                    squared_image_diff          = image_diff**2
                    squared_template_diff       = template_diff**2
                    
                    sum_of_image_template_diff_products = sum_of_image_template_diff_products + image_template_diff_product
                    sum_of_squared_image_diffs          = sum_of_squared_image_diffs          + squared_image_diff
                    sum_of_squared_template_diffs       = sum_of_squared_template_diffs       + squared_template_diff

            #
            # compute final result
            #
            result_value = 0
        
            if sum_of_squared_image_diffs != 0 and sum_of_squared_template_diffs != 0:
                result_value = sum_of_image_template_diff_products / ( ( sum_of_squared_image_diffs * sum_of_squared_template_diffs )**( 0.5 ) )
            
            result[y,x] = result_value

    result_cropped = result[ template_half_height:num_pixels_y-template_half_height, template_half_width:num_pixels_x-template_half_width ]
    
    return result_cropped
```

## call CPU normalized cross-correlation

```python
r_response_cpu_loop = normalized_cross_correlation_cpu_loop(r.astype(numpy.float32), r_template.astype(numpy.float32))
g_response_cpu_loop = normalized_cross_correlation_cpu_loop(g.astype(numpy.float32), g_template.astype(numpy.float32))
b_response_cpu_loop = normalized_cross_correlation_cpu_loop(b.astype(numpy.float32), b_template.astype(numpy.float32))

response_cpu_loop   = r_response_cpu_loop * g_response_cpu_loop * b_response_cpu_loop



figsize(20,4)

matplotlib.pyplot.subplot(141);
matplotlib.pyplot.imshow(r_response_cpu_loop, cmap="gray");
matplotlib.pyplot.title("r_response_cpu_loop");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(142);
matplotlib.pyplot.imshow(g_response_cpu_loop, cmap="gray");
matplotlib.pyplot.title("g_response_cpu_loop");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(143);
matplotlib.pyplot.imshow(b_response_cpu_loop, cmap="gray");
matplotlib.pyplot.title("b_response_cpu_loop");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(144);
matplotlib.pyplot.imshow(response_cpu_loop, cmap="gray");
matplotlib.pyplot.title("response_cpu_loop");
matplotlib.pyplot.colorbar();
```

## verify correctness

```python
diff                     = numpy.abs(response_opencv - response_cpu_loop)
max_diff                 = numpy.ones_like(diff, dtype=numpy.float32)

template_height          = r_template.shape[0]
template_width           = r_template.shape[1]

template_half_height     = (template_height - 1) / 2
template_half_width      = (template_width  - 1) / 2

response_opencv_small    = response_opencv[ 350-template_half_height:450-template_half_height, 200-template_half_width:300-template_half_width ]
response_cpu_loop_small  = response_cpu_loop[ 350-template_half_height:450-template_half_height, 200-template_half_width:300-template_half_width ]
diff_small               = numpy.abs(response_opencv_small - response_cpu_loop_small)



print \
    "Difference between OpenCV and hand-coded CPU loop results as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))
    
print

figsize(24,14)

matplotlib.pyplot.subplot(231);
matplotlib.pyplot.imshow(response_opencv, cmap="gray");
matplotlib.pyplot.title("response_opencv");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(232);
matplotlib.pyplot.imshow(response_cpu_loop, cmap="gray");
matplotlib.pyplot.title("response_cpu_loop");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(233);
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(234);
matplotlib.pyplot.imshow(response_opencv_small, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("response_opencv_small");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(235);
matplotlib.pyplot.imshow(response_cpu_loop_small, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("response_cpu_loop_small");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(236);
matplotlib.pyplot.imshow(diff_small, cmap="gray", interpolation="nearest");
matplotlib.pyplot.title("diff_small");
matplotlib.pyplot.colorbar();
```
