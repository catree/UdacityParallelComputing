# CPU GPU Convolution Speed Test

The purpose of this code is to perform a 2D uniform blur operation using both the GPU and the CPU. We compare the performance of each method using the system timer.

## load image

```python
import PIL
import PIL.Image

image                 = PIL.Image.open("cinque_terre.jpg")
image_array_rgb       = numpy.array(image)

r,g,b                 = numpy.split(image_array_rgb, 3, axis=2)
a                     = numpy.ones_like(r) * 255

image_array_rgba      = numpy.concatenate((r,g,b,a), axis=2).copy()



figsize(6,4)

matplotlib.pyplot.imshow(image_array_rgba);
matplotlib.pyplot.title("image_array_rgba");
```

## isolate small greyscale patch from the original image

```python
r_small = r[1150:1450,2300:2600,0].copy().astype(numpy.uint8)



figsize(4,4)

matplotlib.pyplot.imshow(r_small, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small");
```

## common uniform blur code

```python
uniform_blur_kernel_width      = numpy.int32(9)
uniform_blur_kernel_half_width = numpy.int32(4)
uniform_blur_kernel_value      = numpy.float32(1) / numpy.float32(uniform_blur_kernel_width**2)
```

## CPU uniform blur

```python
import scipy
import scipy.ndimage
import scipy.ndimage.filters

convolution_filter  = numpy.ones((uniform_blur_kernel_width, uniform_blur_kernel_width), dtype=numpy.float32) * uniform_blur_kernel_value
r_small_blurred_cpu = scipy.ndimage.filters.convolve(r_small, convolution_filter, mode="nearest")



figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_cpu, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu");
```

## GPU naive uniform blur

```python
import pycuda.autoinit
import pycuda.driver
import pycuda.compiler

r_small_blurred_gpu_naive = numpy.zeros_like(r_small)

source_module = pycuda.compiler.SourceModule \
(
"""
__global__ void naive_uniform_blur(
    unsigned char* d_blurred,
    unsigned char* d_original,
    int            num_pixels_y,
    int            num_pixels_x,
    int            blur_kernel_half_width,
    float          blur_kernel_value )
{
    int  ny             = num_pixels_y;
    int  nx             = num_pixels_x;
    int2 image_index_2d = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int  image_index_1d = ( nx * image_index_2d.y ) + image_index_2d.x;

    if ( image_index_2d.x < nx && image_index_2d.y < ny )
    {
        float result = 0;
    
        for ( int y = -blur_kernel_half_width; y <= blur_kernel_half_width; y++ )
        {
            for ( int x = -blur_kernel_half_width; x <= blur_kernel_half_width; x++ )
            {
                int2 image_offset_index_2d         = make_int2( image_index_2d.x + x, image_index_2d.y + y );
                int2 image_offset_index_2d_clamped = make_int2( min( nx - 1, max( 0, image_offset_index_2d.x ) ), min( ny - 1, max( 0, image_offset_index_2d.y ) ) );
                int  image_offset_index_1d_clamped = ( nx * image_offset_index_2d_clamped.y ) + image_offset_index_2d_clamped.x;
    
                result += blur_kernel_value * d_original[ image_offset_index_1d_clamped ];
            }
        }
    
        d_blurred[ image_index_1d ] = (unsigned char)result;
    }
}
"""
)

naive_uniform_blur_function       = source_module.get_function("naive_uniform_blur")

num_pixels_y                      = numpy.int32(r_small.shape[0])
num_pixels_x                      = numpy.int32(r_small.shape[1])
naive_uniform_blur_function_block = (32,16,1)

num_blocks_y                      = int(ceil(float(num_pixels_y) / float(naive_uniform_blur_function_block[1])))
num_blocks_x                      = int(ceil(float(num_pixels_x) / float(naive_uniform_blur_function_block[0])))
naive_uniform_blur_function_grid  = (num_blocks_x, num_blocks_y)

r_small_device                    = pycuda.driver.mem_alloc(r_small.nbytes)
r_small_blurred_device            = pycuda.driver.mem_alloc(r_small_blurred_gpu_naive.nbytes)

pycuda.driver.memcpy_htod(r_small_device,         r_small)
pycuda.driver.memcpy_htod(r_small_blurred_device, r_small_blurred_gpu_naive)

naive_uniform_blur_function(
    r_small_blurred_device,
    r_small_device,
    num_pixels_y,
    num_pixels_x,
    uniform_blur_kernel_half_width,
    uniform_blur_kernel_value,
    block=naive_uniform_blur_function_block,
    grid=naive_uniform_blur_function_grid)

pycuda.driver.memcpy_dtoh(r_small_blurred_gpu_naive, r_small_blurred_device)



figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_gpu_naive, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_naive");
```

## GPU shared memory uniform blur

```python
import pycuda.autoinit
import pycuda.driver
import pycuda.compiler

r_small_blurred_gpu_shared_memory = numpy.zeros_like(r_small)

source_module = pycuda.compiler.SourceModule \
(
"""
#define BLOCK_SIZE_Y           8
#define BLOCK_SIZE_X           32
#define BLUR_KERNEL_HALF_WIDTH 4
#define SHARED_MEMORY_SIZE_Y   BLOCK_SIZE_Y + ( 2 * BLUR_KERNEL_HALF_WIDTH )
#define SHARED_MEMORY_SIZE_X   BLOCK_SIZE_X + ( 2 * BLUR_KERNEL_HALF_WIDTH ) + 1
#define SHARED_MEMORY_OFFSET_Y BLUR_KERNEL_HALF_WIDTH
#define SHARED_MEMORY_OFFSET_X BLUR_KERNEL_HALF_WIDTH

__global__ void shared_memory_uniform_blur(
    unsigned char* d_blurred,
    unsigned char* d_original,
    int            num_pixels_y,
    int            num_pixels_x,
    int            blur_kernel_half_width,
    float          blur_kernel_value )
{
    __shared__ unsigned char s_original[ SHARED_MEMORY_SIZE_Y ][ SHARED_MEMORY_SIZE_X ];

    int  ny                            = num_pixels_y;
    int  nx                            = num_pixels_x;
    int2 image_index_2d_global         = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int2 image_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_index_2d_global.x ) ), min( ny - 1, max( 0, image_index_2d_global.y ) ) );
    int  image_index_1d_global_clamped = ( nx * image_index_2d_global_clamped.y ) + image_index_2d_global_clamped.x;
    int2 image_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y );

    //
    // load center of shared memory
    //
    s_original[ image_index_2d_shared_memory.y ][ image_index_2d_shared_memory.x ] = d_original[ image_index_1d_global_clamped ];
    
    //
    // load y+1 halo into shared memory
    //
    if ( threadIdx.y < BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( ( blockIdx.y + 1 ) * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y + BLOCK_SIZE_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load y-1 halo into shared memory
    //
    if ( threadIdx.y >= BLOCK_SIZE_Y - BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( ( blockIdx.y - 1 ) * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y - BLOCK_SIZE_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load x+1 halo into shared memory
    //
    if ( threadIdx.x < BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( ( blockIdx.x + 1 ) * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X + BLOCK_SIZE_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load x-1 halo into shared memory
    //
    if ( threadIdx.x >= BLOCK_SIZE_X - BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( ( blockIdx.x - 1 ) * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X - BLOCK_SIZE_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load x+1,y+1 halo into shared memory
    //
    if ( threadIdx.x < BLUR_KERNEL_HALF_WIDTH && threadIdx.y < BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( ( blockIdx.x + 1 ) * blockDim.x ) + threadIdx.x, ( ( blockIdx.y + 1 ) * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X + BLOCK_SIZE_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y + BLOCK_SIZE_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load x+1,y-1 halo into shared memory
    //
    if ( threadIdx.x < BLUR_KERNEL_HALF_WIDTH && threadIdx.y >= BLOCK_SIZE_Y - BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( ( blockIdx.x + 1 ) * blockDim.x ) + threadIdx.x, ( ( blockIdx.y - 1 ) * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X + BLOCK_SIZE_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y - BLOCK_SIZE_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load x-1,y+1 halo into shared memory
    //
    if ( threadIdx.x >= BLOCK_SIZE_X - BLUR_KERNEL_HALF_WIDTH && threadIdx.y < BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( ( blockIdx.x - 1 ) * blockDim.x ) + threadIdx.x, ( ( blockIdx.y + 1 ) * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X - BLOCK_SIZE_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y + BLOCK_SIZE_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // load x-1,y-1 halo into shared memory
    //
    if ( threadIdx.x >= BLOCK_SIZE_X - BLUR_KERNEL_HALF_WIDTH && threadIdx.y >= BLOCK_SIZE_Y - BLUR_KERNEL_HALF_WIDTH )
    {
        int2 image_halo_index_2d_global         = make_int2( ( ( blockIdx.x - 1 ) * blockDim.x ) + threadIdx.x, ( ( blockIdx.y - 1 ) * blockDim.y ) + threadIdx.y );
        int2 image_halo_index_2d_global_clamped = make_int2( min( nx - 1, max( 0, image_halo_index_2d_global.x ) ), min( ny - 1, max( 0, image_halo_index_2d_global.y ) ) );
        int  image_halo_index_1d_global_clamped = ( nx * image_halo_index_2d_global_clamped.y ) + image_halo_index_2d_global_clamped.x;
        int2 image_halo_index_2d_shared_memory  = make_int2( threadIdx.x + SHARED_MEMORY_OFFSET_X - BLOCK_SIZE_X, threadIdx.y + SHARED_MEMORY_OFFSET_Y - BLOCK_SIZE_Y );

        s_original[ image_halo_index_2d_shared_memory.y ][ image_halo_index_2d_shared_memory.x ] = d_original[ image_halo_index_1d_global_clamped ];
    }

    //
    // wait until all threads in the thread block are finished loading the image chunk into shared memory
    //
    __syncthreads();

    //
    // perform blur operation by reading image from shared memory
    //
    if ( image_index_2d_global.x < nx && image_index_2d_global.y < ny )
    {
        float result = 0;
    
        for ( int y = -blur_kernel_half_width; y <= blur_kernel_half_width; y++ )
        {
            for ( int x = -blur_kernel_half_width; x <= blur_kernel_half_width; x++ )
            {
                int2 image_offset_index_2d = make_int2( image_index_2d_shared_memory.x + x, image_index_2d_shared_memory.y + y );

                result += blur_kernel_value * s_original[ image_offset_index_2d.y ][ image_offset_index_2d.x ];
            }
        }

        d_blurred[ image_index_1d_global_clamped ] = (unsigned char)result;
    }
}
"""
)

shared_memory_uniform_blur_function = source_module.get_function("shared_memory_uniform_blur")

num_pixels_y                              = numpy.int32(r_small.shape[0])
num_pixels_x                              = numpy.int32(r_small.shape[1])
shared_memory_uniform_blur_function_block = (32,8,1)

num_blocks_y                              = int(ceil(float(num_pixels_y) / float(shared_memory_uniform_blur_function_block[1])))
num_blocks_x                              = int(ceil(float(num_pixels_x) / float(shared_memory_uniform_blur_function_block[0])))
shared_memory_uniform_blur_function_grid  = (num_blocks_x, num_blocks_y)

r_small_device                            = pycuda.driver.mem_alloc(r_small.nbytes)
r_small_blurred_device                    = pycuda.driver.mem_alloc(r_small_blurred_gpu_shared_memory.nbytes)

pycuda.driver.memcpy_htod(r_small_device,         r_small)
pycuda.driver.memcpy_htod(r_small_blurred_device, r_small_blurred_gpu_shared_memory)

shared_memory_uniform_blur_function(
    r_small_blurred_device,
    r_small_device,
    num_pixels_y,
    num_pixels_x,
    uniform_blur_kernel_half_width,
    uniform_blur_kernel_value,
    block=shared_memory_uniform_blur_function_block,
    grid=shared_memory_uniform_blur_function_grid)

pycuda.driver.memcpy_dtoh(r_small_blurred_gpu_shared_memory, r_small_blurred_device)



figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_gpu_shared_memory, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_shared_memory");
```

## verify correctness of GPU naive

```python
diff     = numpy.abs(r_small_blurred_cpu.astype(numpy.float32) - r_small_blurred_gpu_naive.astype(numpy.float32))
max_diff = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between CPU and GPU naive results as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_cpu, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_gpu_naive, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_naive")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## verify correctness of GPU shared memory

```python
diff     = numpy.abs(r_small_blurred_gpu_naive.astype(numpy.float32) - r_small_blurred_gpu_shared_memory.astype(numpy.float32))
max_diff = numpy.ones_like(diff, dtype=numpy.float32) * 255



print \
    "Difference between GPU naive and GPU shared memory results as a percentage of the maximum possible difference (should be 0%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))

print

figsize(14,4)

matplotlib.pyplot.subplot(131)
matplotlib.pyplot.imshow(r_small_blurred_gpu_naive, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_naive")

matplotlib.pyplot.subplot(132)
matplotlib.pyplot.imshow(r_small_blurred_gpu_shared_memory, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_shared_memory")

matplotlib.pyplot.subplot(133)
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff")
matplotlib.pyplot.colorbar();
```

## common benchmarking code

```python
import sys
import time

if sys.platform == "win32":
    print "Using time.clock for benchmarking...\n" 
    system_timer_function = time.clock
else:
    print "Using time.time for benchmarking...\n" 
    system_timer_function = time.time
    
num_timing_iterations = 100
    
    
    
print "num_timing_iterations = %d" % num_timing_iterations
```

## benchmark CPU code using the system timer

```python
total_time_seconds = 0
convolution_filter = numpy.ones((uniform_blur_kernel_width, uniform_blur_kernel_width), dtype=numpy.float32) * uniform_blur_kernel_value

for i in range(num_timing_iterations):
    
    r_small_blurred_cpu  = numpy.zeros_like(r_small).squeeze()
    
    start_time_seconds   = system_timer_function()

    r_small_blurred_cpu = scipy.ndimage.filters.convolve(r_small, convolution_filter, mode="nearest")

    end_time_seconds     = system_timer_function()
    elapsed_time_seconds = (end_time_seconds - start_time_seconds)
    total_time_seconds   = total_time_seconds + elapsed_time_seconds

average_time_seconds_cpu = total_time_seconds / num_timing_iterations



print "Using system timer for benchmarking (see above)..."
print "Average time elapsed executing a uniform blur on the CPU over %d runs: %f s" % (num_timing_iterations,average_time_seconds_cpu)

print 

figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_cpu, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_cpu")
```

## benchmark GPU naive code using the system timer

```python
total_time_seconds = 0

r_small_blurred_gpu_naive = numpy.zeros_like(r_small)

for i in range(num_timing_iterations):
    
    pycuda.driver.memcpy_htod(r_small_blurred_device, r_small_blurred_gpu_naive)
    
    pycuda.driver.Context.synchronize()
    
    start_time_seconds = system_timer_function()

    naive_uniform_blur_function(
        r_small_blurred_device,
        r_small_device,
        num_pixels_y,
        num_pixels_x,
        uniform_blur_kernel_half_width,
        uniform_blur_kernel_value,
        block=naive_uniform_blur_function_block,
        grid=naive_uniform_blur_function_grid)

    pycuda.driver.Context.synchronize()
    
    end_time_seconds     = system_timer_function()
    elapsed_time_seconds = end_time_seconds - start_time_seconds
    total_time_seconds   = total_time_seconds + elapsed_time_seconds

pycuda.driver.memcpy_dtoh(r_small_blurred_gpu_naive, r_small_blurred_device)

average_time_seconds_gpu_naive = total_time_seconds / num_timing_iterations



print "Using system timer for benchmarking (see above)..."
print "Average time elapsed executing naive_uniform_blur GPU kernel over %d runs: %f s" % (num_timing_iterations,average_time_seconds_gpu_naive)

print 

figsize(4,4)

matplotlib.pyplot.imshow(r_small_blurred_gpu_naive, cmap="gray", vmin=0, vmax=255);
matplotlib.pyplot.title("r_small_blurred_gpu_naive")
```

## benchmark GPU shared memory code using the system timer
