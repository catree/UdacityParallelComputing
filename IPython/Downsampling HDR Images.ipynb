# Loading and Saving HDR Images

The purpose of this code is demonstrate how to load and save HDR images.

```python
import array
import OpenEXR
import Imath

pt            = Imath.PixelType(Imath.PixelType.FLOAT)
rgb_hdr       = OpenEXR.InputFile("memorial.exr")
header        = rgb_hdr.header()
dataWindow    = header["dataWindow"]
displayWindow = header["displayWindow"]
sz            = (dataWindow.max.x - dataWindow.min.x + 1, dataWindow.max.y - dataWindow.min.y + 1)

r_hdr, g_hdr, b_hdr = rgb_hdr.channels("RGB", pt)

r = numpy.fromstring(r_hdr, dtype=numpy.float32).reshape((sz[1], sz[0]))
g = numpy.fromstring(g_hdr, dtype=numpy.float32).reshape((sz[1], sz[0]))
b = numpy.fromstring(b_hdr, dtype=numpy.float32).reshape((sz[1], sz[0]))

r = r[0:sz[1]:2,0:sz[0]:2]
g = g[0:sz[1]:2,0:sz[0]:2]
b = b[0:sz[1]:2,0:sz[0]:2]

dataWindow.max.y    = dataWindow.max.y / 2
dataWindow.max.x    = dataWindow.max.x / 2
displayWindow.max.y = displayWindow.max.y / 2
displayWindow.max.x = displayWindow.max.x / 2

rgb = numpy.concatenate((r[:,:,newaxis],g[:,:,newaxis],b[:,:,newaxis]), axis=2).copy()

r_save, g_save, b_save = array.array("f"), array.array("f"), array.array("f")

r_save.fromlist(list(r.ravel()))
g_save.fromlist(list(g.ravel()))
b_save.fromlist(list(b.ravel()))

r_save_string, g_save_string, b_save_string = [ color_channel.tostring() for color_channel in (r_save, g_save, b_save) ]
header["channels"]                          = { "R" : Imath.Channel(pt, 1, 1), "G" : Imath.Channel(pt, 1, 1), "B" : Imath.Channel(pt, 1, 1) }
save_file                                   = OpenEXR.OutputFile("memorial_small.exr", header)

save_file.writePixels({'R' : r_save_string, 'G' : g_save_string, 'B' : b_save_string })



figsize(14, 14)

matplotlib.pyplot.imshow(rgb);
matplotlib.pyplot.title("rgb");
```
